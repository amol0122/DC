```python
import time

def round_robin(tasks, servers):
    num_servers = len(servers)
    server_index = 0
    distribution = {}

    for task in tasks:
        print(f"\nReceived request from: {task}")
        server = servers[server_index]
        distribution.setdefault(server, []).append(task)
        server_index = (server_index + 1) % num_servers

        time.sleep(1)
        print(f"'{task}' assigned to Server: '{server}'")
        time.sleep(0.5)

    return distribution

# List of servers
servers = ['192.168.1.101', '192.168.1.102', '192.168.1.103']

# Simulated tasks/requests
tasks = ['Task1', 'Task2', 'Task3', 'Task4', 'Task5', 'Task6', 'Task7', 'Task8', 'Task9', 'Task10']

# Load balancing using round-robin algorithm
distribution = round_robin(tasks, servers)

# Print the distribution of tasks among servers
print("\nFinal Distribution:")

for server, tasks_assigned in distribution.items():
    print(f"Server {server}: {', '.join(tasks_assigned)}")
```
